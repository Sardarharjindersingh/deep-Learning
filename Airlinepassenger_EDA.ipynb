```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
sns.set(rc={'figure.figsize':(11.7,8.27)})
import os
```

```python
df = pd.read_csv(r"C:\Users\sarda\OneDrive\Desktop\airline_passenger_satisfaction.csv")
df.head()
```

```python
df.describe()
```

```python
chart = sns.countplot(x="Age", data=df)
for ind, label in enumerate(chart.get_xticklabels()):
    if int(label.get_text()) % 10 == 0:
        label.set_visible(True)
    else:
        label.set_visible(False)
```

```python
fig, ax = plt.subplots(1,2)
sns.histplot(x="Gender", data=df, stat="density", shrink=0.8, color="steelblue", ax=ax[0])
sns.histplot(x="Customer Type", data=df, stat="density", shrink=0.8, color="steelblue",ax=ax[1])
fig.show()
```

```python
num_list = list(df.columns)

fig = plt.figure(figsize=(10,30))

for i in range(len(num_list)):
    plt.subplot(8,3,i+1)
    plt.title(num_list[i])
    plt.xticks(rotation=45)
    plt.hist(df[num_list[i]],color='blue',alpha=0.5)

plt.tight_layout()
```

```python
plt.figure(figsize = (15,8))
sns.heatmap(df.corr(),annot=True,fmt='.1f')
```
