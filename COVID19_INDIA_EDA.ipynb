```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True , axis = 1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active cases
covid_df['Active Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
covid_df['State/UnionTerritory'] = covid_df['State/UnionTerritory'].replace(['Maharashtra***','Karanataka','Madhya Pradesh***','Bihar****','Himanchal Pradesh'],['Maharashtra','Karnataka','Madhya Pradesh','Bihar','Himachal Pradesh'])
```

```python
Statewise = pd.pivot_table(covid_df, values = ["Confirmed" , "Deaths" , "Cured"], index = "State/UnionTerritory" , aggfunc = max)
```

```python
Statewise["Recovery rate"] = Statewise["Cured"]*100 / Statewise["Confirmed"]
```

```python
Statewise["Mortality rate"] = Statewise["Deaths"]*100 / Statewise["Confirmed"]
```

```python
Statewise = Statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
Statewise.style.background_gradient(cmap='cubehelix')
```

```python
#Top 10 states by Active cases 
top_10_Active_Cases=covid_df.groupby(by = "State/UnionTerritory").max()[["Active Cases","Date"]].sort_values(by=["Active Cases"],ascending= False).reset_index()
fig = plt.figure(figsize=(16,8))
plt.title("Top 10 states with most Active Cases in India", size=25)
ax = sns.barplot(data = top_10_Active_Cases.iloc[:10],y="Active Cases", x="State/UnionTerritory", linewidth= 2,edgecolor="red",)
plt.xlabel('States')
plt.ylabel('Active Cases')
plt.show()
```

```python
#Top 10 states by high Mortality rate 
top_10_Mortality_rate = Statewise.groupby(by = "State/UnionTerritory").max()[["Mortality rate"]].sort_values(by=["Mortality rate"],ascending= False ).reset_index()
fig = plt.figure(figsize=(16,8))
plt.title("Top 10 states with most Mortality rate in India", size=30)
ax = sns.barplot(data = top_10_Mortality_rate.iloc[:10],y="Mortality rate", x="State/UnionTerritory", linewidth= 2,edgecolor="red",)
plt.xlabel('States')
plt.ylabel('Mortality rate')
plt.show()
```

```python
#Top 10 states by high Recovery rate 
top_10_Confirmed_Cases = covid_df.groupby(by = "State/UnionTerritory").max()[["Confirmed"]].sort_values(by=["Confirmed"],ascending= False ).reset_index()
fig = plt.figure(figsize=(16,8))
plt.title("Top 10 states with Most Confirmed Cases in India", size=30)
ax = sns.barplot(data = top_10_Confirmed_Cases.iloc[:10],y="Confirmed", x="State/UnionTerritory", linewidth= 2,edgecolor="red",)
plt.xlabel('States')
plt.ylabel('Confirmed')
plt.show()
```

```python
#Top 10 states by Death Reported 
top_10_Death_Reported = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths"]].sort_values(by=["Deaths"],ascending= False ).reset_index()
fig = plt.figure(figsize=(16,5))
plt.title("Top 10 states with Most Deaths Reported in India", size=30)
ax = sns.barplot(data = top_10_Death_Reported.iloc[:10],y="Deaths", x="State/UnionTerritory", linewidth= 2,edgecolor="red",)
plt.xlabel('States')
plt.ylabel('Deaths')
plt.show()
```

```python
#Top 10 states by high Recovery rate 
top_10_Confirmed_Cases = covid_df.groupby(by = "State/UnionTerritory").max()[["Confirmed"]].sort_values(by=["Confirmed"],ascending= False ).reset_index()
fig = plt.figure(figsize=(16,8))
plt.title("Top 10 states with Most Confirmed Cases in India", size=30)
ax = sns.lineplot(data = top_10_Confirmed_Cases.iloc[:10],y="Confirmed", x="State/UnionTerritory", linewidth= 5)
plt.xlabel('States')
plt.ylabel('Confirmed')
plt.show()
```

```python
#Covid cases growth trend by year
fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Rajasthan','Kerala','Karnataka','Tamil Nadu'])], x = 'Date',y = 'Active Cases',hue = 'State/UnionTerritory')
ax.set_title("Covid cases growth trend by year in India", size=16)
```

```python
#Male vs Female Vaccination
Male = vaccination["Male(Individuals Vaccinated)"].sum()
Female = vaccination["Female(Individuals Vaccinated)"].sum()
data = [Male,Female]
fig = plt.figure(figsize =(10, 7))
plt.pie(data,labels = ["Male","Female"],autopct='%1.1f%%',startangle=180,shadow= True,explode=(0,0.05,))
plt.legend()
plt.title("Male vs Female Vaccination across India", size=16)
```
